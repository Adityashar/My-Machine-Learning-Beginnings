```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.array([1,2,3,4,5,6])
ys = np.array([5,4,6,5,6,7])
plt.scatter(xs, ys)
plt.show()
```

```python
m = ((xs * ys).mean() - xs.mean() * ys.mean())/((xs ** 2).mean() - xs.mean() ** 2)
b = ys.mean() - m * xs.mean()
m, b
```

```python
xtest = np.array([0, 1, 3, 6])
ypred = m*xtest + b
plt.scatter(xs, ys, Color = 'b')
plt.plot(xtest, ypred, Color = 'g', marker = '^')
plt.show()
```
