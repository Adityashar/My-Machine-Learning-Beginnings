```python
import matplotlib.pyplot as plt
```

```python
x = [5,8,10]
y = [10,16,20]

plt.plot(x, y, Color = 'r', label = "val 1")
plt.scatter(x, y)

x1 = [15,18,20]
y1 = [12,16,20]

plt.plot(x1, y1, 'r--')
#plt.plot(x1, y1, Color = 'r', marker = '^', label = "val 2")

plt.axis([0,25, 0, 30])


plt.grid(True)

plt.legend()


#plt.text(5, 5, "abc", fontsize = 100)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Demo")
plt.show()
```
